```python
#Lets nill the null attendance with average of attendance


avg_attendance = world_cup.Attendance.dropna().sum() / len(world_cup.Attendance.dropna())

print(avg_attendance)
```

```python
world_cup.Attendance = world_cup.Attendance.fillna(avg_attendance)
world_cup.isnull().sum() #now the sum null values is zero
```

```python
#Now change the columns Attendance and Goals to float to int. Beacause then it makes more sense

#we will do it the following way:

world_cup['Home Team Goals'] = world_cup['Home Team Goals'].astype('int64')
world_cup['Away Team Goals'] = world_cup['Away Team Goals'].astype('int64')
world_cup['Half-time Home Goals'] = world_cup['Half-time Home Goals'].astype('int64')
world_cup['Half-time Away Goals'] = world_cup['Half-time Away Goals'].astype('int64')
world_cup['Attendance'] = world_cup['Attendance'].astype('int64')
world_cup.dtypes

```

```python
#Lets to convert the column Datetime to datetime

world_cup.Datetime = pd.to_datetime(world_cup.Datetime)
world_cup.dtypes
```

```python
#Okay. Our dataset is now adequently clean. Lets to do some analysis!
```

```python
#Many people say it's much harder to score goals nowadays than it was a long time ago. 
##############IS IT TRUE?

#Are there more or less goals scored in the 2014 World Cup than there was in the another World Cups?


```

```python
world_cup['Total Goals'] = world_cup['Home Team Goals'] + world_cup['Away Team Goals']

world_cup.head()
```

```python
world_cup.tail()
```

```python
world_cup.shape
```

```python
grouped_by_year = world_cup.groupby('Year')
goals_by_year = grouped_by_year['Total Goals'].sum()
goals_by_year.plot(color = 'red', figsize = (10,7))
plt.title('Total Goals Scored', fontsize = 18)
plt.xlabel('World Cup')
plt.ylabel('Goals');

plt.show()
```

```python
games_per_year = world_cup['Year'].value_counts()
games_per_year.sort_index().plot(color = 'red', figsize = (10,7));
plt.title('Total Games Played', fontsize = 18)
plt.xlabel('World Cup')
plt.ylabel(' Games');

plt.show()
```

```python
goals_per_game_per_year = goals_by_year/games_per_year
goals_per_year
```

```python
goals_per_game_per_year.plot(figsize = (10,7), color = 'Teal')
plt.title('Average Goals per Game by Year', fontsize = 20)
plt.xlabel('World Cup')
plt.ylabel('Average Goals per Game')
plt.show()
```

```python
world_cup['Hour Played'] = round(world_cup['Datetime'].dt.minute/60 + world_cup['Datetime'].dt.hour)
```

```python
grouped_by_time_played = world_cup.groupby('Hour Played')
avg_goal_per_hour = grouped_by_time_played['Total Goals'].sum()/grouped_by_time_played['Total Goals'].count()
avg_goal_per_hour.plot(kind = 'bar', figsize = (18,9), color = 'green');
plt.title('Goals Scored at Times of Day', fontsize=20)
plt.xlabel('Hour of the day', fontsize=14)
plt.ylabel('Goals', fontsize = 14);
plt.show()
```

```python
goals_by_home_team = world_cup[['Home Team Goals', 'Home Team Initials']].groupby('Home Team Initials')
ax = goals_by_home_team.sum() \
.sort_values('Home Team Goals', ascending = False) \
.head(25) \
.plot(kind ='bar', figsize =(10,8))

goals_by_away_team = world_cup[['Away Team Goals', 'Away Team Initials']].groupby('Away Team Initials')
goals_by_away_team.sum() \
.sort_values('Away Team Goals', ascending = False) \
.head(25) \
.plot(kind = 'bar', figsize=(10,8), ax=ax, stacked = True, color = 'red')

plt.title('Total Goals Scored by Team')
plt.xlabel('Team Initials')
plt.ylabel('Goals')
plt.show()
```
